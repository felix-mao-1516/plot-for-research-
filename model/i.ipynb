```python
# Re-run the meshgrid tutorial (the previous kernel state was reset).

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd


# --- 1) What meshgrid returns ---
x = np.array([0, 1, 2])         # 3 x-values
y = np.array([10, 20])          # 2 y-values

X_xy, Y_xy = np.meshgrid(x, y, indexing="xy")  # Default intuition: rows correspond to y, cols to x

# Display small grids as tables
df_X = pd.DataFrame(X_xy, columns=[f"x{j}" for j in range(X_xy.shape[1])])
df_Y = pd.DataFrame(Y_xy, columns=[f"x{j}" for j in range(Y_xy.shape[1])])


# --- 2) Visualize the grid points ---
plt.figure()
plt.scatter(X_xy, Y_xy)
for i in range(Y_xy.shape[0]):
    for j in range(X_xy.shape[1]):
        plt.text(X_xy[i, j], Y_xy[i, j], f"({int(X_xy[i, j])},{int(Y_xy[i, j])})", ha="center", va="bottom", fontsize=8)
plt.title("Grid points produced by meshgrid(indexing='xy')")
plt.xlabel("X")
plt.ylabel("Y")
plt.tight_layout()
plt.show()

print(X_xy, Y_xy)

```
